```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

fig = plt.figure(figsize=(14, 10))

# ===== رسم 3D =====
ax1 = fig.add_subplot(121, projection='3d')

def draw_box(ax, x, y, z, dx, dy, dz, color='cyan', alpha=0.4):
    vertices = [
        [x,y,z], [x+dx,y,z], [x+dx,y+dy,z], [x,y+dy,z],
        [x,y,z+dz], [x+dx,y,z+dz], [x+dx,y+dy,z+dz], [x,y+dy,z+dz]
    ]
    faces = [
        [vertices[0],vertices[1],vertices[2],vertices[3]],
        [vertices[4],vertices[5],vertices[6],vertices[7]],
        [vertices[0],vertices[1],vertices[5],vertices[4]],
        [vertices[2],vertices[3],vertices[7],vertices[6]],
        [vertices[0],vertices[3],vertices[7],vertices[4]],
        [vertices[1],vertices[2],vertices[6],vertices[5]]
    ]
    poly = Poly3DCollection(faces, alpha=alpha, edgecolor='black', linewidth=1.5, facecolor=color)
    ax.add_collection3d(poly)

# الجسم
draw_box(ax1, 0,0,0, 50,50,60, color='burlywood', alpha=0.3)
# الدرج
draw_box(ax1, 2,2,44, 46,46,14, color='peru', alpha=0.7)
# مقبض الدرج
draw_box(ax1, 22,0,49, 6,2,4, color='silver', alpha=0.9)
# الباب
draw_box(ax1, 2,0,2, 46,2,38, color='saddlebrown', alpha=0.7)
# مقبض الباب
draw_box(ax1, 22,0,20, 6,2,4, color='silver', alpha=0.9)

ax1.set_xlim(0,55)
ax1.set_ylim(0,55)
ax1.set_zlim(0,65)
ax1.set_xlabel('العرض (cm)')
ax1.set_ylabel('العمق (cm)')
ax1.set_zlabel('الارتفاع (cm)')
ax1.set_title('مجر 3D', fontsize=14, fontweight='bold')

# ===== رسم 2D أمامي =====
ax2 = fig.add_subplot(122)
ax2.set_xlim(-10, 70)
ax2.set_ylim(-10, 75)
ax2.set_aspect('equal')

# الإطار الخارجي
rect = plt.Rectangle((0,0), 50, 60, fill=False, edgecolor='black', linewidth=2)
ax2.add_patch(rect)

# الدرج
drawer = plt.Rectangle((2,44), 46, 14, fill=True, facecolor='peru', edgecolor='black', linewidth=1.5)
ax2.add_patch(drawer)
ax2.plot([25],[51], 'o', color='silver', markersize=10)  # مقبض

# الباب
door = plt.Rectangle((2,2), 46, 38, fill=True, facecolor='saddlebrown', edgecolor='black', linewidth=1.5, alpha=0.7)
ax2.add_patch(door)
ax2.plot([25],[20], 'o', color='silver', markersize=10)  # مقبض

# القياسات
ax2.annotate('', xy=(55,0), xytext=(55,60),
            arrowprops=dict(arrowstyle='<->', color='red', lw=2))
ax2.text(58, 30, '60 cm', color='red', fontsize=11, va='center')

ax2.annotate('', xy=(0,-7), xytext=(50,-7),
            arrowprops=dict(arrowstyle='<->', color='blue', lw=2))
ax2.text(25, -10, '50 cm', color='blue', fontsize=11, ha='center')

ax2.annotate('', xy=(-7,44), xytext=(-7,58),
            arrowprops=dict(arrowstyle='<->', color='green', lw=2))
ax2.text(-9, 51, 'درج\n14cm', color='green', fontsize=9, ha='center')

ax2.annotate('', xy=(-7,2), xytext=(-7,40),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax2.text(-9, 21, 'باب\n38cm', color='purple', fontsize=9, ha='center')

ax2.set_title('واجهة أمامية', fontsize=14, fontweight='bold')
ax2.axis('off')

plt.suptitle('مجر - 50×50×60 cm', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('/Users/user/Desktop/majer_plan.png', dpi=150, bbox_inches='tight')
plt.show()
print("تم الحفظ على Desktop!")
```
