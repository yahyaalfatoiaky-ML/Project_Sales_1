````python
import { useState } from "react";

const categories = [
  { id: "university", label: "🎓 الجامعة", color: "#4F46E5" },
  { id: "ml", label: "🤖 ML", color: "#059669" },
  { id: "english", label: "🇬🇧 English", color: "#DC2626" },
  { id: "books", label: "📚 كتب", color: "#D97706" },
  { id: "other", label: "✅ أخرى", color: "#7C3AED" },
];

const initialTasks = [
  { id: 1, text: "كمّل Week 1 ديال Coursera", category: "ml", done: false, priority: "high" },
  { id: 2, text: "اقرا العادة الأولى", category: "books", done: false, priority: "medium" },
  { id: 3, text: "تعلم 5 كلمات إنجليزية", category: "english", done: false, priority: "low" },
  { id: 4, text: "راجع دروس الجامعة", category: "university", done: false, priority: "high" },
  { id: 5, text: "افتح الفصل 1 من Hands-on ML", category: "books", done: false, priority: "medium" },
];

const priorities = {
  high: { label: "مهم جداً", color: "#EF4444", bg: "#FEF2F2" },
  medium: { label: "متوسط", color: "#F59E0B", bg: "#FFFBEB" },
  low: { label: "عادي", color: "#10B981", bg: "#F0FDF4" },
};

export default function TodoApp() {
  const [tasks, setTasks] = useState(initialTasks);
  const [newTask, setNewTask] = useState("");
  const [selectedCat, setSelectedCat] = useState("ml");
  const [selectedPriority, setSelectedPriority] = useState("medium");
  const [filter, setFilter] = useState("all");
  const [nextId, setNextId] = useState(6);

  const addTask = () => {
    if (!newTask.trim()) return;
    setTasks([...tasks, {
      id: nextId,
      text: newTask,
      category: selectedCat,
      done: false,
      priority: selectedPriority,
    }]);
    setNextId(nextId + 1);
    setNewTask("");
  };

  const toggleTask = (id) => {
    setTasks(tasks.map(t => t.id === id ? { ...t, done: !t.done } : t));
  };

  const deleteTask = (id) => {
    setTasks(tasks.filter(t => t.id !== id));
  };

  const filtered = filter === "all" ? tasks
    : filter === "done" ? tasks.filter(t => t.done)
    : filter === "pending" ? tasks.filter(t => !t.done)
    : tasks.filter(t => t.category === filter);

  const done = tasks.filter(t => t.done).length;
  const progress = tasks.length ? Math.round((done / tasks.length) * 100) : 0;

  return (
    <div style={{
      minHeight: "100vh",
      background: "linear-gradient(135deg, #0f0c29, #302b63, #24243e)",
      fontFamily: "'Segoe UI', sans-serif",
      padding: "20px",
      direction: "rtl",
    }}>
      <div style={{ maxWidth: 600, margin: "0 auto" }}>

        {/* Header */}
        <div style={{ textAlign: "center", marginBottom: 28 }}>
          <h1 style={{ color: "#fff", fontSize: 28, fontWeight: 800, margin: 0 }}>
            📋 To-Do ديالك
          </h1>
          <p style={{ color: "#a5b4fc", margin: "6px 0 0" }}>ML · جامعة · English · كتب</p>
        </div>

        {/* Progress */}
        <div style={{
          background: "rgba(255,255,255,0.08)",
          borderRadius: 16,
          padding: "16px 20px",
          marginBottom: 20,
          backdropFilter: "blur(10px)",
        }}>
          <div style={{ display: "flex", justifyContent: "space-between", marginBottom: 8 }}>
            <span style={{ color: "#e0e7ff", fontWeight: 600 }}>التقدم اليومي</span>
            <span style={{ color: "#a5b4fc" }}>{done}/{tasks.length} مهام</span>
          </div>
          <div style={{ background: "rgba(255,255,255,0.15)", borderRadius: 99, height: 10 }}>
            <div style={{
              width: `${progress}%`,
              height: "100%",
              borderRadius: 99,
              background: "linear-gradient(90deg, #6366f1, #8b5cf6)",
              transition: "width 0.5s ease",
            }} />
          </div>
          <div style={{ textAlign: "center", color: "#c7d2fe", fontSize: 13, marginTop: 6 }}>
            {progress}% مكتمل 🎯
          </div>
        </div>

        {/* Add Task */}
        <div style={{
          background: "rgba(255,255,255,0.08)",
          borderRadius: 16,
          padding: 16,
          marginBottom: 20,
          backdropFilter: "blur(10px)",
        }}>
          <input
            value={newTask}
            onChange={e => setNewTask(e.target.value)}
            onKeyDown={e => e.key === "Enter" && addTask()}
            placeholder="زيد مهمة جديدة..."
            style={{
              width: "100%",
              background: "rgba(255,255,255,0.1)",
              border: "1px solid rgba(255,255,255,0.2)",
              borderRadius: 10,
              padding: "10px 14px",
              color: "#fff",
              fontSize: 15,
              marginBottom: 10,
              boxSizing: "border-box",
              outline: "none",
            }}
          />

          {/* Category Select */}
          <div style={{ display: "flex", gap: 6, flexWrap: "wrap", marginBottom: 10 }}>
            {categories.map(cat => (
              <button key={cat.id} onClick={() => setSelectedCat(cat.id)} style={{
                padding: "5px 10px",
                borderRadius: 99,
                border: "none",
                cursor: "pointer",
                fontSize: 12,
                fontWeight: 600,
                background: selectedCat === cat.id ? cat.color : "rgba(255,255,255,0.1)",
                color: "#fff",
                transition: "all 0.2s",
              }}>{cat.label}</button>
            ))}
          </div>

          {/* Priority Select */}
          <div style={{ display: "flex", gap: 6, marginBottom: 12 }}>
            {Object.entries(priorities).map(([key, val]) => (
              <button key={key} onClick={() => setSelectedPriority(key)} style={{
                padding: "5px 12px",
                borderRadius: 99,
                border: "none",
                cursor: "pointer",
                fontSize: 12,
                fontWeight: 600,
                background: selectedPriority === key ? val.color : "rgba(255,255,255,0.1)",
                color: "#fff",
                transition: "all 0.2s",
              }}>{val.label}</button>
            ))}
          </div>

          <button onClick={addTask} style={{
            width: "100%",
            padding: "11px",
            borderRadius: 10,
            border: "none",
            cursor: "pointer",
            background: "linear-gradient(135deg, #6366f1, #8b5cf6)",
            color: "#fff",
            fontWeight: 700,
            fontSize: 15,
          }}>+ زيد المهمة</button>
        </div>

        {/* Filter */}
        <div style={{ display: "flex", gap: 6, flexWrap: "wrap", marginBottom: 16 }}>
          {[
            { id: "all", label: "الكل" },
            { id: "pending", label: "⏳ باقي" },
            { id: "done", label: "✅ مكتمل" },
            ...categories,
          ].map(f => (
            <button key={f.id} onClick={() => setFilter(f.id)} style={{
              padding: "5px 12px",
              borderRadius: 99,
              border: "none",
              cursor: "pointer",
              fontSize: 12,
              fontWeight: 600,
              background: filter === f.id ? "#6366f1" : "rgba(255,255,255,0.1)",
              color: "#fff",
              transition: "all 0.2s",
            }}>{f.label}</button>
          ))}
        </div>

        {/* Tasks */}
        <div style={{ display: "flex", flexDirection: "column", gap: 8 }}>
          {filtered.length === 0 && (
            <div style={{ textAlign: "center", color: "#a5b4fc", padding: 30 }}>
              🎉 ما كاين والو دابا!
            </div>
          )}
          {filtered.map(task => {
            const cat = categories.find(c => c.id === task.category);
            const pri = priorities[task.priority];
            return (
              <div key={task.id} style={{
                background: task.done ? "rgba(255,255,255,0.04)" : "rgba(255,255,255,0.09)",
                borderRadius: 12,
                padding: "12px 16px",
                display: "flex",
                alignItems: "center",
                gap: 12,
                backdropFilter: "blur(10px)",
                borderRight: `4px solid ${cat?.color}`,
                opacity: task.done ? 0.6 : 1,
                transition: "all 0.3s",
              }}>
                <input
                  type="checkbox"
                  checked={task.done}
                  onChange={() => toggleTask(task.id)}
                  style={{ width: 18, height: 18, cursor: "pointer", accentColor: "#6366f1" }}
                />
                <div style={{ flex: 1 }}>
                  <div style={{
                    color: task.done ? "#9ca3af" : "#f0f4ff",
                    textDecoration: task.done ? "line-through" : "none",
                    fontSize: 14,
                    fontWeight: 500,
                  }}>{task.text}</div>
                  <div style={{ display: "flex", gap: 6, marginTop: 4 }}>
                    <span style={{
                      fontSize: 11,
                      background: `${cat?.color}33`,
                      color: cat?.color,
                      padding: "2px 8px",
                      borderRadius: 99,
                      fontWeight: 600,
                    }}>{cat?.label}</span>
                    <span style={{
                      fontSize: 11,
                      background: `${pri?.color}33`,
                      color: pri?.color,
                      padding: "2px 8px",
                      borderRadius: 99,
                      fontWeight: 600,
                    }}>{pri?.label}</span>
                  </div>
                </div>
                <button onClick={() => deleteTask(task.id)} style={{
                  background: "transparent",
                  border: "none",
                  cursor: "pointer",
                  color: "#ef4444",
                  fontSize: 16,
                  padding: "2px 6px",
                }}>🗑️</button>
              </div>
            );
          })}
        </div>

        {done === tasks.length && tasks.length > 0 && (
          <div style={{
            textAlign: "center",
            marginTop: 24,
            padding: 16,
            background: "rgba(16,185,129,0.15)",
            borderRadius: 12,
            color: "#6ee7b7",
            fontWeight: 700,
            fontSize: 16,
          }}>
            🎉 بركا عليك! كملتي كل المهام اليوم! 💪
          </div>
        )}

        <div style={{ textAlign: "center", color: "#6366f1", marginTop: 20, fontSize: 12 }}>
          كل يوم خطوة → ML + جامعة + English 🚀
        </div>
      </div>
    </div>
  );
}
```

## To-Do ديالك جاهز! 🎉

---

## شنو فيه؟ ✅
```
🎓 فئات → جامعة, ML, English, كتب
🔴 أولوية → مهم جداً, متوسط, عادي
📊 Progress Bar → كيبين قداش كملتي
🔍 فلتر → كتشوف حسب الفئة
🗑️ حذف → تمسح أي مهمة
````
