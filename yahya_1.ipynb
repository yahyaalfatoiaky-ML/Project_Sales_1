```python
def process_cpu(text):
    text = str(text).lower()
    
    # Intel High-End
    if 'i7' in text or 'i9' in text or 'ultra 7' in text or 'ultra 9' in text:
        return 'Intel High-End'
    # Intel Mid-Range
    elif 'i5' in text or 'ultra 5' in text:
        return 'Intel Mid-Range'
    # Intel Entry-Level
    elif 'i3' in text:
        return 'Intel i3'
    # AMD High-End
    elif 'ryzen 7' in text or 'ryzen 9' in text:
        return 'AMD High-End'
    # AMD Mid-Range
    elif 'ryzen 5' in text:
        return 'AMD Ryzen 5'
    # AMD Entry-Level
    elif 'ryzen 3' in text:
        return 'AMD Ryzen 3'
    # Budget Processors (Celeron, Pentium, etc.)
    elif 'celeron' in text or 'pentium' in text or 'athlon' in text:
        return 'Budget CPU'
    else:
        return 'Other'

# تطبيق الدالة على العمود
data['cpu_brand'] = data['processor'].apply(process_cpu)
```

```python
data = pd.get_dummies(data,columns=["cpu_brand"] , prefix="cpu")
data.head()
```

```python
def process_os(text):
    text = str(text).lower()
    if "windows" in text:
        return "windows"
    elif "Mac" in text:
        return "Mac"
    elif "DOS" in text:
        return "DOS/No OS"
    else:
        return 'Others'
    
data["os_type"] = data["os"].apply(process_os)
data = data[data["os_type"] != "Others"]

    
```

```python
def simplify_cpu(text):
    text = str(text).lower()
    if 'i3' in text: return 'i3'
    if 'i5' in text: return 'i5'
    if 'i7' in text: return 'i7'
    if 'i9' in text: return 'i9'
    if 'ryzen 3' in text: return 'Ryzen 3'
    if 'ryzen 5' in text: return 'Ryzen 5'
    if 'ryzen 7' in text: return 'Ryzen 7'
    if 'ryzen 9' in text: return 'Ryzen 9'
    if 'm1' in text or 'm2' in text or 'm3' in text: return 'Apple Silicon'
    return 'Other'

# كنطبقو هاد الدالة على العمود
data['cpu_brand'] = data['processor'].apply(simplify_cpu)

# دابا نحولو هاد الماركات الجديدة لأرقام (One-Hot Encoding)
data = pd.get_dummies(data, columns=['cpu_brand'], prefix='cpu')

# ونمسحو عمود processor الأصلي
data.drop('processor', axis=1, inplace=True)
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# 1. تحميل الداتا
df = pd.read_csv('Lap 2.csv')

# 2. تحويل النصوص لأرقام (ضروري باش الموديل يخدم)
df_encoded = pd.get_dummies(df)

# 3. فصل "الثمن" عن "المواصفات"
X = df_encoded.drop('price', axis=1) 
y = np.log1p(df_encoded['price']) # باش نحسبو اللوغاريتم

# 4. السكالينغ (الميزان)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. تدريب الموديل
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_scaled, y)

# 6. حفظ الملفات (هنا فين غيبانو الملفات اللي مسحتي دابا)
joblib.dump(model, 'laptop_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(X.columns.tolist(), 'model_columns.pkl')

print("✅ الملفات تساوبو دابا! عاود دير 'ls' في Terminal غتلقاهم.")
```
