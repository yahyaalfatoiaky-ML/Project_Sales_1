```python
import matplotlib.pyplot as plt

# بيانات تجريبية
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 9, 6]

plt.scatter(x, y, color='blue') # رسم النقاط
plt.title("Scatter Plot using Matplotlib") # إضافة عنوان
plt.xlabel("X Axis") # تسمية المحور الأفقي
plt.ylabel("Y Axis") # تسمية المحور الرأسي
plt.grid(True) # إضافة شبكة خلفية يدوياً
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# بيانات تجريبية
data = {"X Axis": [1, 2, 3, 4, 5], "Y Axis": [5, 7, 4, 9, 6]}

# رسم النقاط بلمسة واحدة (تضيف التنسيق والشبكة وتسمية المحاور تلقائياً)
sns.scatterplot(data=data, x="X Axis", y="Y Axis")

plt.title("Scatter Plot using Seaborn")
plt.show()
```
